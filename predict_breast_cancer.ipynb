```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer_dataset = load_breast_cancer()
print(breast_cancer_dataset)
```

```python
dataset = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
print(dataset)
```

```python
x = dataset.drop(columns = 'label', axis = 1)
y = dataset['label']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
crmat = classification_report(y_test, y_predict)
cmat = confusion_matrix(y_test, y_predict)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
plt.figure(figsize = (7,5))
plt.xlabel('Predicted')
plt.ylabel('Actual')
sb.heatmap(cmat,annot = True)
```

```python
from sklearn.metrics import accuracy_score
#training accuraccy
y_train_predict = model.predict(x_train)
train_data_accurracy = accuracy_score(y_train, y_train_predict)
print('Training Accuracy: ',train_data_accurracy)
```
