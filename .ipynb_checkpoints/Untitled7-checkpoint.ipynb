```python
#Allow to send HTTP request
import requests
#fetching data from world wide web
import urllib.request
import time
#for parsing HTML/XML documents
from bs4 import BeautifulSoup
```

```python
#copy the url of the webpage
url = "https://www.simplilearn.com/big-data-and-analytics/senior-data-scientist-masters-program-training"
```

```python
#send a request to that url
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text,"html.parser")
```

```python
soup
```

```python
content = soup.find("div",{"class":"exam-body"})
```

```python
content
```

```python
article = ''
for i in content.findAll('p'):
    article = article+" "+i.text
print(article)
```
