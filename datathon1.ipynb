```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Distribution of Purchase Amount
plt.figure(figsize=(10, 6))
sns.histplot(dataset['Purchase Amount (USD)'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Purchase Amount')
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of transaction frequency
plt.figure(figsize=(10, 6))
sns.countplot(x='Frequency of Purchases', data=dataset, palette='pastel')
plt.title('Distribution of Transaction Frequency')
plt.xlabel('Transaction Frequency')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of selling products
plt.figure(figsize=(10, 6))
sns.countplot(x='Item Purchased', data=dataset, palette='pastel')
plt.title('top selling products')
plt.xlabel('selling products')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of selling products
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=dataset, palette='pastel')
plt.title('top selling Category of products')
plt.xlabel('category of selling products')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_purchase_amount = dataset.groupby('Gender')['Purchase Amount (USD)'].mean().reset_index()
avg_purchase_amount
```

```python
plt.bar(avg_purchase_amount['Gender'], avg_purchase_amount['Purchase Amount (USD)'], color = 'yellow')
plt.title('customer purchasing behavior')
plt.xlabel('Gender')
plt.ylabel('Purchase_Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
dfMale = dataset[dataset['Gender'] == 'Male']
```

```python
dfMale
```

```python
avg_purchase_amount_male = dfMale.groupby('Category')['Purchase Amount (USD)'].mean().reset_index()
plt.bar(avg_purchase_amount_male['Category'], avg_purchase_amount_male['Purchase Amount (USD)'], color = 'orange')
plt.title('purchasing behavior of customer(Male)')
plt.xlabel('Category')
plt.ylabel('Purchase_Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
dfFemale = dataset[dataset['Gender'] == 'Female']
dfFemale
```

```python
avg_purchase_amount_female = dfFemale.groupby('Category')['Purchase Amount (USD)'].mean().reset_index()
plt.bar(avg_purchase_amount_female['Category'], avg_purchase_amount_female['Purchase Amount (USD)'], color = 'orange')
plt.title('purchasing behavior of customer(Female)')
plt.xlabel('Category')
plt.ylabel('Purchase_Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_purchase_amount_female = dfFemale.groupby('Item Purchased')['Previous Purchases'].mean().reset_index()

plt.bar(avg_purchase_amount_female['Item Purchased'], avg_purchase_amount_female['Previous Purchases'], color='orange')
plt.title('purchasing behavior of customer(Femle)')
plt.xlabel('Item Purchased')
plt.ylabel('Previous Purchases')
plt.xticks(rotation=90)
plt.show()
```

```python
avg_purchase_amount_male = dfMale.groupby('Item Purchased')['Previous Purchases'].mean().reset_index()

plt.bar(avg_purchase_amount_male['Item Purchased'], avg_purchase_amount_male['Previous Purchases'], color='orange')
plt.title('purchasing behavior of customer(Male)')
plt.xlabel('Item Purchased',fontsize = 8)
plt.ylabel('Previous Purchases')
plt.xticks(rotation=90)
plt.show()
```
